```python
import os
import random

def delete_half_files(directory):
    # Get a list of all files in the specified directory
    files = [file for file in os.listdir(directory) if os.path.isfile(os.path.join(directory, file))]
    
    # Shuffle the list of files randomly
    random.shuffle(files)
    
    # Calculate the number of files to keep (half of total)
    num_to_keep = 500
    
    # Iterate over the files and delete half of them
    for file in files[num_to_keep:]:
        file_path = os.path.join(directory, file)
        try:
            # Delete the file
            os.unlink(file_path)
            print(f"Deleted file: {file_path}")
        except Exception as e:
            print(f"Failed to delete {file_path}: {e}")

# Specify the directory where you want to delete half of the files
for i in range(20):
    directory_to_modify = "data/" + str(i)
    delete_half_files(directory_to_modify)
```

```python
import os

def count_files_in_folder(directory):
    # Initialize a counter for files
    file_count = 0
    
    # Iterate over all items in the directory
    for item in os.listdir(directory):
        # Check if the item is a file
        if os.path.isfile(os.path.join(directory, item)):
            # Increment the file counter
            file_count += 1
    
    return file_count

# Specify the directory for which you want to count the files
for i in range(20):
    directory_to_count = f"data/{i}"  # Change this to your desired directory

# Call the function to count the files in the specified directory
    num_files = count_files_in_folder(directory_to_count)
    print(f"Number of files in {directory_to_count}: {num_files} {i}")
```
